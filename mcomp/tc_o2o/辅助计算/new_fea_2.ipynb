```python
import sys 
sys.path.append('../app/')
import fileio
import preprocess
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import featureprocess

import time
import threading
```

```python
# 读取数据
dfoff ,dftest,dfon = fileio.read_all()

# 特征处理
dfoff = preprocess.get_label(dfoff)

#dfoff = preprocess.process_manjian(dfoff)

#两个特征，一个是um，一个是uc,,我觉得um靠谱一点
dfoff.loc[:,'UM_id'] = dfoff.User_id.astype('str')+'_'+dfoff.Merchant_id.astype('str')

#这次使用未泄露的数据,
dfoff,_ = preprocess.split_receivetime(dfoff)

```

```python
user_off_list = dfoff.UM_id.unique()
```

```python
def user_features(df001):
    """返回userid，只领券，核销数，直接买，单商铺最大购买次数，成交商家数"""
    lcdict = {0:0,1:0,-1:0}
    #df001 = dfoff[dfoff.UM_id==uid]
    
    dflabelcount = df001.label.value_counts()
    dfMerchantcount = df001.Merchant_id.value_counts()
    for i in dflabelcount.index:
        lcdict[i] = dflabelcount[i]
    
    return [lcdict[0],lcdict[1],lcdict[-1],dfMerchantcount.max(),dfMerchantcount.count()]
```

```python
time_start=time.time()
ll = []
n = 0
for i in dfoff.groupby('UM_id'):
    
    if n%1000==0:
        print("---{0}%".format(n*100/len(user_off_list)) )
    ll.append([i[0]]+user_features(i[1]))
    
    n = n+1

time_end=time.time()    
print('time cost',time_end-time_start,'s')
```

```python
dfobj = pd.DataFrame(ll)
dfobj.index = range(0,len(dfobj))
dfobj.columns = ['用户id','只领券','核销数','直接买','单商铺最大购买次数','成交商家数']
```
