```python
# 获取股票代码和名称
sl = mongoCli('CodeAndName','shares')
x = sl.col.find({},{'_id':0})

CNdict = {}
for kv in  x:
    for key in kv.keys():
            CNdict[key] = kv[key]
```

```python
dd = []
for share in CNdict:
    # print(share) 
    sl = mongoCli('sharePrice',share)
    x = sl.col.find({},{'date':1})  
    l1 = []
    for i in x:
        l1.append(i['date'][0])
    
    df = pd.DataFrame(pd.value_counts(l1),columns=[share])
    df.sort_index(inplace=True)
    dd.append(df)
```

```python
cc = pd.concat(dd,axis=1)
```
