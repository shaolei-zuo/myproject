```python
import pandas as pd
import urllib3
import datetime
import time
import re
```

```python
url = 'https://tp.wjx.top/wjx/join/tpresult.aspx?activity=91423735'
http = urllib3.PoolManager()
r = http.request('GET', url)    
#r.data.decode()
#index = int(time.time())
name_list = re.findall('\d*组.*?家庭',r.data.decode())
```

```python
#now_time = datetime.datetime.now()
#now_time.timestamp()
#time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(t))
#url = 'https://tp.wjx.top/wjx/join/tpresult.aspx?activity=91423735'
dfobj = pd.DataFrame(columns=name_list)

#http = urllib3.PoolManager()
#r = http.request('GET', url)
```

```python
while True:
    
    http = urllib3.PoolManager()
    r = http.request('GET', url)    
    #r.data.decode()
    index = int(time.time())
    name_list = re.findall('\d*组.*?家庭',r.data.decode())
    num_list = [int(i[1:].split('&')[0]) for i in re.findall('>\d*&nbsp;',r.data.decode())]
    for a,b in zip(name_list,num_list):
        #print(a,b)
        dfobj.loc[index,a] = b
    dfobj.to_csv('obj.csv')
    time.sleep(60)
```
