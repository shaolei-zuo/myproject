```python
# 代理数量限制导致有的网站没有收到数据，这里梳理一下
import re
import numpy as np
import pandas as pd
```

```python
with open('./urllog.txt') as f:
    cc = f.readlines()
```

```python
ll = []
for i in cc:
    ee = re.findall('--http://market.finance.sina.com.cn/transHis\.php\?symbol=sz000001&date=(.+?)&page=(.+?)数据储存完成\n',i)
    if ee:
        ll.append(ee)
```

```python
mydic = {}
for i in ll:
    a = i[0][0]
    b = i[0][1]
    if a in mydic.keys():
        mydic[a].append(int(b))
    else:
        mydic[a] = []
        mydic[a].append(int(b)) 
```

```python
for i in mydic.keys():
    this_len = len(mydic[i])
    this_max = max(mydic[i])
    print('{},已获取{}，最大页码{}'.format(i,this_len,this_max))
```
