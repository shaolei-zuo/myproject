# 账单统计

```python
file = './data/alipay_record_20210107_1330_1.csv'
df = pd.read_csv(file,encoding='ANSI',sep=',')
```

```python
df.columns  = [col.replace(' ','') for col in df.columns]
df.资金状态 = [col.replace(' ','') for col in df.资金状态]
```

```python
df.iloc[:3,5::].head(3)
```

```python
df["time"] = pd.to_datetime(df['交易创建时间'])
```

```python
df.loc[:,'month']=[i.month for i in df.loc[:,'time'] ]
```

```python
df.loc[:,'class']= df.loc[:,'交易对方'].replace(class_dict)#交易分类
dfzy = df[df.资金状态 == '资金转移']
dfzc = df[df.资金状态 == '已支出']#支出，包括买基金什么的
dfsr = df[df.资金状态 == '已收入']
dfno = df[df.资金状态 == '']
dfczc = dfzc[dfzc['class']!='未知']#去掉买基金之类的投资交易
dfczc2 = dfczc.drop(dfczc[(dfczc['class']=='交通') & (dfczc.loc[:,'金额（元）']>150)].index)#去掉大笔的交通支出，主要是公务或者机票
```

```python
dd = []
for i in dfczc2.groupby(['month']):
    cc = pd.DataFrame(i[1].groupby('class')['金额（元）'].sum())
    cc.columns = [i[0]]
    dd.append(cc)
```

```python
dflast = pd.concat(dd,axis=1,sort=False)
dflast = dflast.iloc[:,-3:]
dflast.loc[:,'平均'] = dflast.sum(axis=1)/len(dflast.columns)
dflast.loc['sum',:] = dflast.sum()
dflast
```

```python
dflast.to_csv('cc.csv')
```
